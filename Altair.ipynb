```python
df=pd.read_excel("C:/Users/Cristofer/Bootcamp_python/Semana_3/MusicData.xlsx")
df
```

# Uso los 5 primeros formato de venta que mas han vendido

```python
df2=df[(df["Metric"]=="Value (Adjusted)") ][["Format","Value (Actual)"]].groupby("Format").sum().sort_values("Value (Actual)", ascending=False).head(5)
```

## Filtro los datos y solo uso la metria de valor ajustado

```python
lista=list(df2.index)
df=df[df["Metric"]=="Value (Adjusted)"]
df=df[df["Format"].isin(lista)]
df
```

## Creo el grafico con altair


```python
multi= alt.selection_multi()
click = alt.selection_multi(encodings=['color'])
chart =alt.Chart(df).mark_point().encode(
    x=alt.X('Year:N'),
    y='Value (Actual):Q',
    shape="Format",
    tooltip=["Format","Year","Value (Actual)"],
    color=alt.condition(multi,"Format",alt.value("lightgray"))
).properties( selection=multi , width=800,height=500)
chart
```

## Guardo el archivo el formato html 

```python
chart.save('Musica.html')
```
